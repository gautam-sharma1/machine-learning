```python
def replace_zeros_w_random_normal(DF,label, mu, sigma):
    truth_1 = DF[label] == 0
    random = np.random.normal(mu, sigma, DF.shape[0])
    filt = DF[DF[label] > 0]
    vector_1 = truth_1 * random
    truth_2 = vector_1 == 0
    vector_2 = truth_2 * DF[label]
    DF[label] = np.maximum(vector_1,vector_2)
    return DF

data3 = replace_zeros_w_random_normal(data3,'Wind Speed Daily AVG' ,15.97,6)
data3.describe()
data3.to_csv('Final_weather.csv')
```
